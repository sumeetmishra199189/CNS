```python
import os
import pandas as pd
```

```python
################ MORTALITY DATA #####################

fname = "../data/NCHS_-_Drug_Poisoning_Mortality_by_State__United_States.csv" # Assumes you are in 'code' and 'data' is another folder in parent directory
with open (fname) as f:
    mortality_df = pd.read_csv(f)
```

```python
#Cleaning and aggregating the mortality data by State and Year 
"""
Original Mortality_df
mortality_df.info()
<class 'pandas.core.frame.DataFrame'>
RangeIndex: 2862 entries, 0 to 2861
"""
#Limiting to Our years:
mortality_df = mortality_df.query('Year >= 2006 and Year <=2016 and State != "United States"')

#After Limiting:
mortality_df.info()
"""
Data columns (total 19 columns):
State                                           1749 non-null object
Year                                            1749 non-null int64
Sex                                             1749 non-null object
Age Group                                       1749 non-null object
Race and Hispanic Origin                        1749 non-null object
Deaths                                          1749 non-null int64
Population                                      1749 non-null int64
Crude Death Rate                                1749 non-null float64
Standard Error for Crude Rate                   1749 non-null float64
Lower Confidence Limit for Crude Rate           1749 non-null float64
Upper Confidence Limit for Crude Rate           1749 non-null float64
Age-adjusted Rate                               693 non-null float64
Standard Error for Age-adjusted Rate            693 non-null float64
Lower Confidence Limit for Age-adjusted Rate    693 non-null float64
Upper Confidence Limit for Age-adjusted Rate    693 non-null float64
State Crude Rate in Range                       693 non-null object
US Crude Rate                                   1749 non-null float64
US Age-adjusted Rate                            1749 non-null float64
Unit                                            1749 non-null object
"""
```

```python
####################PRESCRIPTION DATA#####################
#Importing the prescription DF
prescription_fname = "../data/merged_cdc_prescription_2006_2016.csv"
with open (prescription_fname) as f:
    prescription_df = pd.read_csv(f, index_col=0)
prescription_df.head()
```

```python
#Merging the mortality and prescription DFs
full_df = pd.merge(mortality_df, prescription_df, on=['State', 'Year'])

#Checking out Alabama
full_df[full_df.State=='Alabama']
```
